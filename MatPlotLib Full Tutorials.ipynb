```python
# matplotlib is the most important library for Data Visualization
# matplotlib.pyplot is used for plotting graphs
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

# %matplotlib notebook... for interactive plots
# %matplotlib inline... for static plots
%matplotlib inline
```

```python
# generating sin() graph
x=np.linspace(0,5*np.pi,100)   # value of sin() ranges from 0 to pie

# Defining the wave
sin_y=np.sin(x)
cos_y=np.cos(x)

# plotting the wave
plt.plot(x,sin_y,label="sine")
plt.plot(x,cos_y,label="cosine")

plt.legend(loc="lower left")   # displays label in graph and the location of the label in the graph
plt.title("Sin Wave")   # gives title to the graph
plt.show()    # displaying the plots
```

### Different types of plots

```python
# Line Plot... makes a line between 

x=np.linspace(0,4*np.pi,100)   # value of sin() ranges from 0 to pie
y=np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x,y)   # single plot means 1 line

plt.title("Sin and Cosine Wave")   # gives title to the graph

plt.show()
```

```python
# Multi-Line Plot... makes a line between 

x=np.linspace(0,4*np.pi,100)   # value of sin() ranges from 0 to pie
sin_y=np.sin(x)
cos_y=np.cos(x)

plt.figure(figsize=(6,1))
plt.style.use('Solarize_Light2')

# multiple plots means multiple lines    # defining the color and width of the line in the graph...default lineWidth is 1
plt.plot(x, sin_y, '*', color='red', linewidth=4)   # plot-1  
plt.plot(x, cos_y, '--', color='blue', linewidth=2)   # plot-2

plt.title("Sin and Cosine Wave")   # gives title to the graph
plt.legend()
plt.show()
```

```python
# 4 types of linestyle commands....solid, dashed, dotted and dashDot

plt.figure(figsize=(5,2))

x=np.linspace(0,9,100,dtype='int')   # value of sin() ranges from 0 to pie

plt.plot(x, x+0,linestyle='solid', label='SOLID')    # plotting graph with (x-axis, y-axis)
plt.plot(x, x+1, linestyle='dotted', label='DOTTED')
plt.plot(x, x+2, linestyle='dashed', label='DASHED') 
plt.plot(x, x+3, linestyle='dashdot', label='DASHDOT') 

plt.legend(loc='upper left')     # to show the labels on the graph
plt.show()

```

```python
# adjusting the plot axis (fitting the plot)....setting limits for x and y axis

x=np.linspace(0,4*np.pi,100)   # value of sin() ranges from 0 to pie

plt.xlim(0,20)  # Range of x-axis from 0 to 20
plt.ylim(-10,20)  # Range of y-axis from 0 to 20

sin_y=np.sin(x)

plt.plot(x, sin_y) 

plt.title("Sin Wave")   # gives title to the graph

plt.show()
```

```python
# can adjust the graph for best fit by using plt.axis('tight') and plt.axis('equal')
sin_y=np.sin(x)

plt.axis('equal')  # or axis('tight')
plt.plot(x, sin_y) 

plt.title("Sin Wave")   # gives title to the graph

plt.show()
```

```python
# set() combines all the commands of the graph fitting

x=np.linspace(0,4*np.pi,100)   # value of sin() ranges from 0 to pie
fig=plt.axes()
fig.plot(x, np.sin(x))

fig.set(xlim=(0,10), ylim=(-2,2), xlabel='x-axis', ylabel='y-axis', title='Sin Wave')
sin_y=np.sin(x)
plt.plot(x, sin_y) 
plt.title("Sin Wave")   # gives title to the graph
plt.show()
```

### Scatter Plots


```python
# scatter plots are points in 2-D plane
# scatter() is used to create scatter plots
n=100
x=np.random.randn(n)
y=np.random.randn(n)

plt.scatter(x,y,c='black',alpha=1,s=1)   # c is color, alpha is the transparency, s is the size of the dots
plt.title('Scatter Plot')
plt.show()
```

```python
#

n=1000
x1=np.random.randn(n)
y1=np.random.randn(n)

x2=np.random.randn(n)+2    # +2 scatters the points by 2-pixels apart in x-axis
y2=np.random.randn(n)+2    # +2 scatters the points by 2-pixels apart in y-axis

plt.scatter(x1,y1,c='red',alpha=.7,s=10, label='Group1')   # c is color, alpha is the transparency, s is the size of the dots
plt.scatter(x2,y2,c='blue',alpha=.7,s=10, label='Group1')   # label is the name of the group
plt.title('Scatter Plot with Multiple Groups')

plt.show()
```

```python
# applying scatter plots on real life dataSets (skilearn.dataSets)
from sklearn.datasets import load_iris

iris=load_iris()

x=iris.data
y=iris.target

plt.scatter(x[:, 0:1],x[:,1:2],c='red',alpha=.9,s=50, label='Group1')   # scatter plot between 0th cloumn and 1st column
plt.show()
```

```python
# marker() is used fort the shape by which the plot has been made
# marksrs are 'o','x','.',',','+','^','<','>','s','d'
plt.scatter(x[:, 0:1],x[:,1:2],c='red',alpha=.9,s=50, marker='+')   # scatter plot between 0th cloumn and 1st column
plt.show()
```

```python
# generating a random scatter plot
r=np.random.RandomState(0)

x=r.randn(100)
y=r.randn(100)

color=r.rand(100)      # 100 colors   random colors
size=r.rand(100)*1000  # making bigger numbers by multiplication  random sizes

plt.title('Random Scatter Plot')
plt.xlabel('x-values')   # text on x-axis
plt.ylabel('y-values')    # text on y-axis
plt.scatter(x,y,c=color, s=size, alpha=1, cmap='viridis', edgecolors='red')
plt.colorbar()    # adds a color bar at the right of graph making it easy to understand
```

```python
iris=load_iris()
feature=iris.data

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.scatter(feature[:,0], feature[:,1], alpha=0.8, s=feature[:,3]*100, c=iris.target, cmap='viridis')

plt.colorbar()
plt.show()
```

## HISTOGRAMS and BINNINGS

```python
# Histograms are used to display the set of continious or discrete data
# binning is the process of dividing the data intu these intervals (see plot)
data=np.random.randn(100)
plt.hist(data)
plt.show()
```

```python
# changing number if bins (nitervals in x-axis)
data=np.random.randn(100000)

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.title("Random Histogram")
plt.hist(data,bins=2000, edgecolor='red',color='blue', alpha=0.8)
plt.show()
```

```python
# histtype='stepfilled' is used to remove individual bins
data=np.random.randn(1000)

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.title("Random Histogram")

plt.hist(data,bins=20, edgecolor='red',color='blue', alpha=0.8, histtype='stepfilled')
plt.show()
```

```python
# Multiple histograms
x1=np.random.randn(1000)
x2=np.random.randn(1000)+3
x3=np.random.randn(1000)+6

plt.hist(x1,bins=20, alpha=0.8, histtype='stepfilled')
plt.hist(x2,bins=20, alpha=0.8, histtype='stepfilled')
plt.hist(x3,bins=20, alpha=0.8, histtype='stepfilled')


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title("Random Histogram")

plt.show()
```

```python
# Plotting 2-D histogram (top-view like a 3-D pyramid)... shows relationship between 2 variables
x=np.random.randn(1000)
y=np.random.randn(1000)

plt.hist2d(x, y, bins=20, cmap='viridis')


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("2-D Histogram")
plt.colorbar()

plt.show()
```

## PIE-CHARTS

```python
# pie-chart is a circular statistical graph which is divided into slices
cars=['BMW', 'AUDI', 'TESLA','CRETA', 'THAR']
data=[450,    354,      53,      100,    10]

plt.pie(data, labels=cars)
plt.show()
```

## BASIC OPERATIONS ON PIE-CHARTS

```python
# slicing out a specific part.... 'explode=" "'

cars=['BMW', 'AUDI', 'TESLA','CRETA', 'THAR']
data=[450,    354,      53,      100,    10]
splitOut=[0,0,0,.1,0]
plt.pie(data, labels=cars, explode=splitOut)
plt.show()
```

```python
# adding shadows  (shodow=True)
cars=['BMW', 'AUDI', 'TESLA','CRETA', 'THAR']
data=[450,    354,      53,      100,    10]
splitOut=[0,0,0,.1,0]
plt.pie(data, labels=cars, explode=splitOut, shadow='True')
plt.show()
```

```python
# adding peprcentages
cars=['BMW', 'AUDI', 'TESLA','CRETA', 'THAR']
data=[450,    354,      53,      100,    30]
splitOut=[0,0,0,.1,0]
plt.pie(data, labels=cars, explode=splitOut, autopct='%1.f%%')
plt.show()
```

```python
# adding wedget
cars=['BMW', 'AUDI', 'TESLA','CRETA', 'THAR']
data=[450,    354,      53,      100,    30]
splitOut=[0,0,0,.1,0]
plt.pie(data, labels=cars, explode=splitOut, autopct='%1.f%%',
                  wedgeprops={ 'edgecolor' : 'black',   # color of the borders
                               'linewidth' : 2,         # width of the borders
                               'antialiased': False})   # degrades the quality of plot by reducing pixels
plt.show()
```

## Index Based Multiple SubPlots

```python
# means many plots in a single graph
x=np.linspace(0,10,50)
fig,ax=plt.subplots(2,2)   # creates graphs in 2*2 matrix (total 4 graphs in a single plot)

y=np.sin(x)    # plot 1
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')

y=np.cos(x)    # plot 2
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y=np.tan(x)    # plot 3
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')

y=np.random.randn(1000)    # plot 4
ax[1,1].hist(y, bins=40)
ax[1,1].set_title('Random Histogram')

plt.tight_layout()
plt.show()
```

## 3D Plots

```python
# 3-D Plots has 3-axis (x,y,z)
plt.axes(projection='3d')
```

```python
# giving values in 3D plot
ax=plt.axes(projection = '3d')

z=np.linspace(0,15,10)
y=np.cos(z)
x=np.sin(z)

ax.plot3D(x,y,z, 'blue')
```

```python
ax=plt.axes(projection = '3d')

z=np.linspace(0,15,100)
y=np.cos(z)
x=np.sin(z)

ax.plot3D(x,y,z, 'blue')
```

```python
# another way to make a 3-D plot
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')

x=np.random.randn(100)
y=np.random.randn(100)
z=np.random.randn(100)

ax.scatter(x, y, z, c='blue', marker='*')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# plt.tight_layout()
plt.show()
```
