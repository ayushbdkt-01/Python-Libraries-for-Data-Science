```python
# Slicing in DataFrame using loc() and iloc()
data={"one"     : pd.Series([10,20,30,40]),
      "two"     : pd.Series([100,200,300,400]),
      "three" : pd.Series([1000,2000,3000,4000]),
      "four"    : pd.Series([10000,20000,30000,40000])}
df=pd.DataFrame(data)
print(df)

# loc() is based on names of row and columns
# loc( rowStart:rowEnd , columnStart:columnEnd ) is used to select specific rows and columns
df.loc[0:2,'one':'two']   # upperBound is inclusive (included)

# iloc() is based on indexing of row and columns
# iloc( rowStart:rowEnd , columnStart:columnEnd ) is used to select specific rows and columns
df.iloc[0:2 , 0:2] # upperBound is exclusive (not included)
df.iloc[[1,3] , [1,2]]   # both syntax of iloc() are same
```

```python
# slicing dataFrame using conditions
df[df["two"]>200]

df.loc[ df["two"]>200 , ["three","four"] ]
```

```python
# adding column in a dataFrame
l=[1,2,3,4]
df["NewCol"]=l
print(df)
print(50*'-')
l1=[1,2,3,4]
df["LatestColumn"]=l1
print(df)
```

```python
# operations on an existing column

df2=df['NewCol']+10
print(df2)
df["FinalColumn"]=df['NewCol']+10
print(70*'-')
print(df)
```

```python
# deleting a column from a dataFrame
# both del and pop() are used to delete a column from the dataFrame
print(df)
print(50*'-')
del df["FinalColumn"]
print(df)
print(50*'-')
df.pop("LatestColumn")
print(df)
```

```python
# adding new rows in the dataFrame
df1=pd.DataFrame( [[1,2],[3,4]] , ['a','b'] )
df2=pd.DataFrame( [[5,6],[7,8]] , ['a','b'] )
print(df1)
print(30*'-')
print(df2)

df3=df1.append(df2)
print(df3)
```

```python
# drop() means to remove a row and column
data={"one"     : pd.Series([10,20,30,40]),
      "two"     : pd.Series([100,200,300,400]),
      "three" : pd.Series([1000,2000,3000,4000]),
      "four"    : pd.Series([10000,20000,30000,40000])}
df=pd.DataFrame(data)
print(df)
print(30*'-')
df1=df.drop([0,1],axis=0)   # removes rowIndex 0 and 1 from the dataFrame (axis=0 --> row)
df2=df.drop(["one","two"],axis=1)   # removes columnIndex 'one' and 'two' from the dataFrame (axis=1 --> column)
print(df1)
print(30*'-')
print(df2)
```

```python
# apply() is used to apply a function to a single element or a whole dataFrame
print(df.apply(np.mean))
print(30*'-')
print(df.apply(np.min))
print(30*'-')
print(df.apply(np.max))
print(30*'-')
print(df.apply(lambda x: x.max()-x.min()))
```

```python
# reindex() is used to change the label (indexing) of row and column
data={"one"     : pd.Series([10,20,30,40]),
      "two"     : pd.Series([100,200,300,400]),
      "three"   : pd.Series([1000,2000,3000,4000]),
      "four"    : pd.Series([10000,20000,30000,40000])}
df=pd.DataFrame(data)
print(df)
print(20*'-')
print(df.reindex([1,0,3,2]))    # row labels are changed
print(20*'-')
print(df.reindex(columns=["four","one","three","two"]))    # columns labels are changed
```

```python
# rename() is used to rename columns
data={"one"     : pd.Series([10,20,30,40]),
      "two"     : pd.Series([100,200,300,400]),
      "three"   : pd.Series([1000,2000,3000,4000]),
      "four"    : pd.Series([10000,20000,30000,40000])}
df=pd.DataFrame(data)
print(df)
print(30*'-')
# columns={} changes column names
# index={} changes row names
df.rename(columns={"one":"ONE","two":"TWO","three":"THREE","four":"FOUR"},inplace=True,index={0:"Z",1:"O",2:"T",3:"T"})   # 'inplace=True' makes the changes in the original data
print(df)
```

```python
# sorting dataFrames
data={"one"     : pd.Series([10,20,30,40]),
      "two"     : pd.Series([100,200,300,400]),
      "three"   : pd.Series([1000,2000,3000,4000]),
      "four"    : pd.Series([10000,20000,30000,40000])}
df=pd.DataFrame(data)
print(df)
print(30*'-')
df.sort_values(by="one")   # sorts dataFrame wrt a specific column 
df.sort_values(by="one",ascending=False)   # sorts dataFrame wrt a specific column in descending order
df.sort_values(by="one",kind='heapsort')   # sorts dataFrame wrt a specific column 
```
